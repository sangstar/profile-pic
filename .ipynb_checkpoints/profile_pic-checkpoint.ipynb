```python
from numpy import sin, cos, arccos, arcsin
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
import matplotlib.animation as animation

class SpringPendulum:
    def __init__(self,
                init_state = [85,0,1,0],
                L = 1.0,
                M = 1.0,
                g = 9.81,
                k = 20.0,
                origin = (0,0)):
        
        self.init_state = np.asarray(init_state, dtype = 'float')
        self.params = (L,M,g,k)
        self.origin = origin
        self.elapsed_time = 0
        self.state = [self.init_state[0] * np.pi / 180., self.init_state[1] * np.pi / 180., 
                      self.init_state[2], self.init_state[3]] 

        
    def position(self):
        (L,M,g,k) = self.params
        
        x = np.cumsum([self.origin[0], self.state[2] * sin(self.state[0])])
        y = np.cumsum([self.origin[1], -self.state[2] * cos(self.state[0])])
        
        return (x,y)
    
    def equation(self, state, t):
        (L,M,g,k) = self.params
        theta = state[0]
        omega = state[1]
        r = state[2]
        drdt = state[3]
        dydt = [omega, -1/r*(g * sin(theta) + 2*omega*drdt), drdt, -(k*(r-L))/ M - g*cos(theta) + r*omega**2]
        return dydt
    
    def time_advance(self, dt):
        self.state = integrate.odeint(self.equation,self.state,[0, dt])[1]
        self.elapsed_time += dt
```

```python
pend = SpringPendulum()
```

```python
fig = plt.figure(figsize=(16,12))
l = pend.params[0]
ax = fig.add_subplot(111, aspect='equal', autoscale_on=False,
                     xlim=(-2*l, 2*l), ylim=(-2*l, 2*l))



```

```python
def generate_plot_points(init_state):
    pend = SpringPendulum(init_state)
    dt = 1/60
    line = []

    for i in range(int(dt*1e5)):
        pend.time_advance(dt)
        line.append(pend.position())



    line = [(subline[0][1], subline[1][1]) for subline in line]


    xs = [val[0] for val in line]
    ys = [val[1] for val in line]
    return xs, ys
```

```python
data1 = generate_plot_points([50,0,1,0])
data2 = generate_plot_points([80,0,1,0])
```

```python
for i in range(10):
    data = generate_plot_points([np.random.randint(i,10)*10,0,1,0])
    plt.plot(data[0],data[1],alpha = 0.05*i)
plt.axis('off')
plt.savefig('propic.png', bbox_inches = 'tight')
```
